### Comparison of Movies and TV Shows Download Rate between United States and India on Netflix

In the world of digital media and entertainment, streaming services like Netflix have become increasingly popular over the years, providing users with a vast array of movies and TV shows at their fingertips. As such, understanding the streaming habits of individuals in different regions is crucial for streaming services to provide tailored content that caters to the preferences of their target audience. This is particularly important in countries like the United States and India, two of the largest markets for digital media, where streaming services can gain a significant share of the market by offering the right mix of content. In this context, recent findings on the download rate and content offerings of Netflix in these countries provide valuable insights that can inform strategic decisions for streaming services and content creators alike.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick
import numpy as np
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df[['month_added', 'month_name_added', 'year_added']] = df['date_added'].apply(lambda x: pd.Series({
'month_added': x.month,
'month_name_added': x.month_name(),
'year_added': x.year
}))

df.head(3)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

countries = ['United States', 'India']
colors = ['blue', 'orange']


for i in range(len(countries)):
    country = countries[i]
    country_rel = df[df['country'] == country]['year_added'].value_counts().sort_index()
    ax.plot(country_rel.index, country_rel, color=colors[i], label=country)
    ax.fill_between(country_rel.index, 0, country_rel, color=colors[i], alpha=0.9)
    
ax.yaxis.tick_right()
ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha = .7)

for s in ['top', 'right','bottom','left']:
    ax.spines[s].set_visible(False)

ax.grid(False)

ax.set_xlim(2008,2020)
plt.xticks(np.arange(2008, 2021, 1))

fig.text(0.13, 0.85, 'United States and India Movies & TV download rate', fontsize=15, fontweight='bold', fontfamily='serif')
fig.text(0.13, 0.59, 
'''

Netflix started slowly adding content over the years, 
but the rate of content additions started to pick up 
in 2015 and increased rapidly from 2016 onwards.

The number of movies and TV shows added over time varies by country, 
with the United States having a higher rate of movie additions and 
India having a higher rate of TV show additions.
'''

, fontsize=12, fontweight='light', fontfamily='serif')


fig.text(0.13,0.2,"United States", fontweight="bold", fontfamily='serif', fontsize=15, color = 'blue')
fig.text(0.266,0.2,"|", fontweight="bold", fontfamily='serif', fontsize=15, color='black')
fig.text(0.279,0.2,"India", fontweight="bold", fontfamily='serif', fontsize=15, color = 'orange')

ax.tick_params(axis=u'both', which=u'both',length=0)

plt.show()
```

### Legend

* Red line and fill represent the movies download rate of United States
* Black line and fill represent the TV shows download rate of United States
* Green line and fill represent the movies download rate of India
* Dark gray line and fill represent the TV shows download rate of India
* X-axis represents the year (2008-2020)
* Y-axis represents the number of movies and TV shows added to Netflix


### Findings

Over time, there has been a rise in the download rate of movies and TV shows on Netflix in both countries.
In the United States, the quantity of movies added to Netflix is greater than that of India, while in India, the quantity of TV shows added is greater. Also, in both nations, the pace of content additions on Netflix has been growing rapidly since 2016.

This information is valuable as it provides insight into the streaming habits of individuals in the United States and India, two of the world's largest markets for digital media. This information can be used by Netflix and other streaming services to improve their content offerings and ensure they are catering to the preferences of their target audience. It may also be useful to content creators who are looking to distribute their work on streaming platforms.
